```python
import os

import re
import pandas as pd
import numpy as np
from helpers import count_words
from tqdm import tqdm
result_dir = "results"
```

```python
all_tweet_files = os.listdir(result_dir)
all_weeks = {}
for i in tqdm(range(53)):
    tweet_files = [
        file_name
        for file_name in all_tweet_files
        if re.findall(rf"2022_{str(i).zfill(2)}_5_\d\d_tweets.csv", file_name)
    ]
    word_count = {}
    for file_name in tweet_files:
        tweets = pd.read_csv(os.path.join(result_dir, file_name)).Text
        count_words(tweets, word_count)
    if word_count:
        all_weeks[i] = word_count
```

```python
words_per_week = pd.DataFrame()

for week, word_count in all_weeks.items():
    words_per_week[week] = word_count

words_per_week = words_per_week.fillna(0)
words_per_week.head()
```

```python
_most_popular_words = words_per_week.max(axis=1).sort_values(ascending=False)[:25]
most_popular_words = _most_popular_words.keys().to_list()
_most_popular_words
```

```python
import plotly.express as px

fig = px.line(words_per_week.transpose(), y=most_popular_words, title="Occurrences of word in tweets per week in 2022")
fig.update_layout(xaxis_title='Week', yaxis_title='Occurrence')
fig.show()
```
