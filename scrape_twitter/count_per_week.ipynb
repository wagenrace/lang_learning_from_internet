```python
import os

import re
import pandas as pd
import numpy as np
from helpers import count_words
from tqdm import tqdm
result_dir = "results"
```

```python
all_tweet_files = os.listdir(result_dir)
all_weeks = {}
for i in tqdm(range(3)):
    tweet_files = [
        file_name
        for file_name in all_tweet_files
        if re.findall(rf"2022_{str(i).zfill(2)}_5_\d\d_tweets.csv", file_name)
    ]
    print(tweet_files)
    word_count = {}
    for file_name in tweet_files:
        tweets = pd.read_csv(os.path.join(result_dir, file_name)).Text
        count_words(tweets, word_count)
    all_weeks[i] = word_count
```

```python
week_df = pd.DataFrame(all_weeks[0].items(), columns=["word", "occurrence"])
number_of_lemmas = np.sum(week_df.occurrence)
week_df.reset_index(drop=True)
week_df = week_df.sort_values("occurrence", ascending=False)
week_df["percentage"] = np.round(week_df.occurrence / number_of_lemmas, 6)
week_df["cumulative"] = week_df.percentage.cumsum()
week_df.to_csv(os.path.join(result_dir, "all_lemmas.csv"), index=False)
week_df.head(n=25)
```

```python
import plotly.graph_objects as go
import urllib, json

url = 'https://raw.githubusercontent.com/plotly/plotly.js/master/test/image/mocks/sankey_energy.json'
response = urllib.request.urlopen(url)
data = json.loads(response.read())

# override gray link colors with 'source' colors
opacity = 0.4
# change 'magenta' to its 'rgba' value to add opacity
data['data'][0]['node']['color'] = ['rgba(255,0,255, 0.8)' if color == "magenta" else color for color in data['data'][0]['node']['color']]
data['data'][0]['link']['color'] = [data['data'][0]['node']['color'][src].replace("0.8", str(opacity))
                                    for src in data['data'][0]['link']['source']]

fig = go.Figure(data=[go.Sankey(
    valueformat = ".0f",
    valuesuffix = "TWh",
    # Define nodes
    node = dict(
      pad = 15,
      thickness = 15,
      line = dict(color = "black", width = 0.5),
      label =  data['data'][0]['node']['label'][:10],
      color =  data['data'][0]['node']['color'][:10]
    ),
    # Add links
    link = dict(
      source =  data['data'][0]['link']['source'][:10],
      target =  data['data'][0]['link']['target'][:10],
      value =  data['data'][0]['link']['value'][:10],
      label =  data['data'][0]['link']['label'][:10],
      color =  data['data'][0]['link']['color'][:10]
))])

fig.update_layout(title_text="Energy forecast for 2050<br>Source: Department of Energy & Climate Change, Tom Counsell via <a href='https://bost.ocks.org/mike/sankey/'>Mike Bostock</a>",
                  font_size=10)
fig.show()
```
